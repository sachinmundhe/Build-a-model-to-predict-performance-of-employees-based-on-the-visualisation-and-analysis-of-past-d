```python
data = data.rename(columns={'KPIs_met >80%': 'KPIs_met', 'awards_won?': 'awards_won'})
data.head()
```

```python
#considering the top 3 factors to visualize the performances in each department
def plots(data):
    plt.figure(figsize=(20,5))
    plt.subplot(1,3,1)
    sns.countplot(data.KPIs_met , hue = data.is_promoted)
    plt.legend(loc = 'best')
    
    plt.subplot(1,3,2)
    sns.countplot(data.awards_won, hue = data.is_promoted)
    plt.legend(loc = 'best')
    
    plt.subplot(1,3,3)     
    sns.countplot(data.avg_training_score, hue = data.is_promoted)
    plt.legend(loc = 'best')
```

```python
Analytics = data[data.department == 'Analytics']
plots(Analytics)
```

```python
Finance = data[data.department == 'Finance']
plots(Finance)
```

```python
HR = data[data.department == 'HR']
plots(HR)
```

```python
Legal = data[data.department == 'Legal']
plots(Legal)
```

```python
Operations = data[data.department == 'Operations']
plots(Operations)
```

```python
R_and_D = data[data.department == 'R&D']
plots(R_and_D)
```

```python
sales_marketing = data[data.department == 'Sales & Marketing']
plots(sales_marketing)
```

```python
Technology = data[data.department == 'Technology']
plots(Technology)
```
